````python
#Detecting Outliers in the data set

# Plot box plots for numerical features
plt.figure(figsize=(16, 12))

# List of numerical columns to check for outliers
numerical_columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
                     'adults', 'children', 'babies', 'adr',
                     'previous_cancellations', 'previous_bookings_not_canceled',
                     'days_in_waiting_list', 'required_car_parking_spaces', 'total_of_special_requests']

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(y=hotel_booking_df[column], color='skyblue')
    plt.title(f'Box Plot of {column}')
    plt.tight_layout()

plt.show()
````

````python
#Dealing with Outliers

# Function to cap outliers at the 1st and 99th percentiles
def cap_outliers(hotel_booking_df, column):
    lower_percentile = hotel_booking_df[column].quantile(0.01)
    upper_percentile = hotel_booking_df[column].quantile(0.99)
    hotel_booking_df[column] = hotel_booking_df[column].clip(lower=lower_percentile, upper=upper_percentile)

# Apply capping to numerical columns with outliers
for column in numerical_columns:
    cap_outliers(hotel_booking_df, column)

# Verify the changes by re-plotting the box plots
plt.figure(figsize=(16, 12))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(y=hotel_booking_df[column], color='lightgreen')
    plt.title(f'Box Plot of {column} After Capping')
    plt.tight_layout()

plt.show()
````
